```python
import pandas as pd
import numpy as np
import re
```

```python
data = pd.read_csv('C:/Users/Sofía Aceves Osuna/ironhack/Clases/attacks.csv')
"""""
Initial plan
1. Clean nans taking in account the most important columns as base to make this decision
2. Choose a relevant index (set 'original order' as index)
3. Clean Date column 
3.1 Homologate format (25-jun-18) changing dtype
3.2 Clean values picking out the most relevant info
4. Turn year into integer( or whatever datatype corresponding to year)
5. Get rid of extra columns ('Unnamed: 22', 'Unnamed: 23')
"""""
data
```

```python
# Get to see and analyse the global perspective of the data
data.info()
```

```python
# Review nulls to see if the decision of eliminating nulls can be made based on the quantity of missing values in each row
data.isnull().sum()
```

```python
# 1. General cleaning: get rid of na
# Reference: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html
# Analysing nulls, noticed that the rows that didn't have data in columns 'Case Number' and 'Date' were empty in the other columns too.
clean_df = data.dropna(axis=0, how='all', thresh= 2, subset = ['Case Number', 'Date'])
```

```python
# 2. Get rid of unuseful columns
# Analysing columns, noticed that the columns 'Unnamed: 22', 'Unnamed: 23' have too much nulls
# Reference: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop.html
final_df = clean_df.drop(columns=['Unnamed: 22', 'Unnamed: 23'])
final_df
# 3. Adjust dtype of column that it's going to turn into the index
type_list = {'original order':int}
final_df1 = final_df.astype(type_list)

#numpy_list = [num for elem in b for elem2 in elem for num in elem2]
# 4. Clean the date column as much as possible, considering the different ways data has been input, using regex
final_df2 = final_df1['Date'] = [re.findall('\d{1,2}\W+\w+\W+\d{4}|\d{2}\W\w+\W\d{2}|\d{4}|\w{3}-\d+', item) for item in final_df['Date']]
final_df1
```

```python
# 5. Set a unique, ordered, and relevant column as index
final_df3 = final_df1.set_index('original order')
```
